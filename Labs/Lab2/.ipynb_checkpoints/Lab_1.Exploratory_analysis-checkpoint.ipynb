```python
data.info()

# Correct IDs duplicates
min_id = min(data["ID"].values)
id_values = data["ID"].value_counts(dropna=False) > 1
bad_id_values = id_values.loc[id_values == True].index.tolist()
for id in bad_id_values:
    indexes_of_bad_ids = data.loc[data["ID"] == id].index.tolist()
    for i in range(len(indexes_of_bad_ids) - 1):
        data.loc[indexes_of_bad_ids[i], "ID"] = min_id - 1
        min_id -= 1
```

```python
# Check for mistakes in catigorical attributes
def print_attributes(types: list) -> None:
    for column in data.columns:
        if data[column].dtypes in types:
            if column == "ID":
                continue
            values = set(data[column].values)
            values_len = len(values)
            print("Attribute: " + column + " "*(10 - len(column)) + "  Values number: " + str(values_len), end="")
            print("  ", values)

print("Catigorical attributes: ")
print_attributes(["object"])
```

```python
# Fix catigorical mistaces
# Attributes with errors:
# - sex
# - pstatus
# - Fedu, Medu
# - Fjob, Mjob
# - guardian
data.loc[data["sex"] == 'm', "sex"] = 'M'
data.loc[data["Pstatus"] == 't', "Pstatus"] = 'T'
data.loc[data["Fedu"] == 'o', "Fedu"] = '0'
data.loc[data["Medu"] == 'o', "Medu"] = '0'
data.loc[data["Fjob"] == 'at-home', "Fjob"] = 'at_home'
data.loc[data["Mjob"] == 'at-home', "Mjob"] = 'at_home'
data.loc[data["guardian"] == 'futher', "guardian"] = 'father'
data["Fedu"] = data["Fedu"].apply(int, convert_dtype=True)
data['Medu'] = data['Medu'].apply(int, convert_dtype=True)

print("Catigorical attributes: ")
print_attributes(["object"])
```

```python
# Check for mistakes in number attributes
print("\n\n\nNumber attributes: ")
print_attributes(["int64", "float64"])

# There no mistakes in values of these attributes
# However there are some nan values
```

### Задание 3  
  
Проанализируйте данные на наличие выбросов.   
Сколько наблюдений можно отнести к выбросам по каждому признаку? Сколько всего выбросов в данных?  
  
**Note:** Выбросы можно не удалять из данных. Это информацию можно учитывать в дальнейшем анализе и удалять выбросы при необходимости. 

```python
# Let's determine outliers in our data
# The function below converts binary attributes to numeric values (0, 1) and other catigorical attributes in numbers, 
# using one-hot encoding. It's will be useful for us soon
def convert_binary_numericDf():
    numericDf = data.copy(deep=True)
    for column in numericDf.columns:
        if numericDf[column].dtype == "object":
            attr_values_num = len(numericDf[column].value_counts())
            attr_values = list(set(numericDf[column].values.tolist()))
            new_values = {attr_values[i]: i for i in range(len(attr_values))}
            print(new_values)
            
            if attr_values_num == 2:
                for index in numericDf[column].index.tolist():
                    numericDf.loc[index, column] = new_values[numericDf.loc[index, column]]
                numericDf[column] = numericDf[column].apply(int, convert_dtype=True)
            else:
                one_hot_columns = pd.get_dummies(numericDf[column], prefix=column)
                idx = numericDf.columns.get_loc(column)
                numericDf.drop(columns=[column], inplace=True)
                for column_ in one_hot_columns:
                    numericDf.insert(idx, column=column_, value=one_hot_columns[column_])
    return numericDf

numericDf = convert_binary_numericDf()
#numericDf
```

```python
# The function below shows outliers in input Data frame, previously converted into numeric data
def show_outliers(data):
    plt.figure(figsize=(15, 10))
    columns_list = [data[column] for column in data.columns if data[column].dtype != "object"]
    columns_names_list = [column for column in data.columns if data[column].dtype != "object"]
    box_plot = plt.boxplot(columns_list)
    fliers_dict = dict()
    for i in range(len(box_plot["fliers"])):
        fliers_data = box_plot["fliers"][i].get_ydata()
        if len(fliers_data) > 0:
            fliers_dict[columns_names_list[i]] = (len(fliers_data), set(fliers_data))
    
    all_outliers = 0
    for key, value in fliers_dict.items():
        print("Attribute: " + '"' + key + '"', " "*(20 - len(key)) + " Number of outliers:",
              value[0], " "*(5 - len(str(value[0]))) + " Outliers: ", value[1])
        all_outliers += value[0]
    print("Number of all outliers: ", all_outliers)
    plt.close()
    
show_outliers(numericDf)
# Outliers, created by boxplots:
```

## Conclusion
* Firstly, according to boxplots, we can conclude, that the values of the math scores are evenly distributed, there are no outliers in G1, G2, and G3.
* As for Portuguese language, there are a lot of outliers in G1 and G2 and not so many in G3. This suggests that the highest density of grade values is in the region of the median and therefore, most people still get good or average grades.
* This fact is also confirmed by the results of density graphs. As you can see, blue lines are more wider than orange lines, that is why Portuguese language students get more good grades, while math students get more different grades.
* Histograms plots shows that show that the number of people who study Portuguese is greater than the number of people who study mathematics

* Хорошо: 14 <= `G3` <= 17
* Удовлетворительно: 8 <= `G3` <= 13
* Неудовлетворительно: `G3` < 8

Оцените распределение нового признака в зависимости от признака `reason` (причина выбора школы). Дайте содержательный комментарий.

```python
newDf = data.copy()
marks = {"Excellent": 4,
         "Good": 3,
         "Satisfactory": 2,
         "Unsatisfactory": 1}

new_column = list()
for g_3 in newDf["G3"]:
    if 18 <= g_3 <= 20:
        new_column.append(marks["Excellent"])
    elif 14 <= g_3 <= 17:
        new_column.append(marks["Good"])
    elif 8 <= g_3 <= 13:
        new_column.append(marks["Satisfactory"])
    elif g_3 < 8:
        new_column.append(marks["Unsatisfactory"])

newDf["marks"] = new_column
newDf
```

```python
temp = newDf[["marks", "reason"]]
temp = temp.groupby("reason").mean()
sns.histplot(data=temp, x="reason", y="marks")
temp
```

## Conclusion
* The average new score was in the range of 2 to 3 (from "Satisfactory" to "Good") for each reason.
* The highest average score is in the reputation category. This suggests that in schools with a good reputation, children learn better than in schools chosen for other reasons
* Lowest average score in the "other" category
* The average score in the "home" category is higher than in the "course" category. This may sound strange, but it seems that children whose schools are nearby are learning better

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier


data_temp = numericDf.copy()
data_temp["marks"] = newDf["marks"]

columns_to_drop = ["marks", "G1", "G2", "G3"]

X = data_temp.drop(columns_to_drop, axis=1)
y = data_temp["marks"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Scale
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# RFC
rfc = RandomForestClassifier(n_estimators=10)
rfc.fit(X_train, y_train)
pred_rfc = rfc.predict(X_test)

print(classification_report(y_test, pred_rfc))
print(confusion_matrix(y_test, pred_rfc))
```
